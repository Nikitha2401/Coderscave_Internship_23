```python
# Assigned the value 0 for the people who are dead and 1 for the people who are not dead

def died(row):
    if row == '9999-99-99':
        return 1 
    else:
        return 0 
df['DIED'] = df.DATE_DIED.apply(died)
df.drop('DATE_DIED',axis=1,inplace=True)
```
